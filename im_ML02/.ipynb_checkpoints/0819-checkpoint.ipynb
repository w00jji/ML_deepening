```python
# 인공신경망 

x1 = [2,4,6,8] # 공부시간 (4명의학생)

y = [20,55,45,80] # 점수

# 부가설명 : 공부 시간이 많으면 점수가 높아지긴하지만 꼭 그렇지만은 않다라는 것을 보여줌

# x1 + 3 : 머신러닝에서 리스트에는 연산이 안된다. 그래서 행렬을 사용해야한다
```

```python
# 인공지능은 행렬(numpy)로 바꿔줘야한다.

# 인공신경망(딥러닝) 학습 원리
import numpy as np
import matplotlib.pyplot as plt
x1 = np.array([2,4,6,8]) # 공부시간 (4명의학생)
x2 = np.array([1,2,1,3]) # 문제집 푼 개수
y = np.array([20,55,45,80]) # 점수

a1 = 0
b = 0
lr =0.005  # 러닝메이트(학습률) = 하이퍼파라미터(사람이 정해주는 것) , 0.001로 했을 때 기울기가 0이 안나왔는데 0.005했을 땐 기울기가 0이 나온다
for i in range(3000): # 미분 3000번 할 거다.
    y_hat = a1*x1+b # y_hat 예측
    #print(y_hat)
    error = y_hat-y # 오차
    #print(error)
    a1_diff = sum(2*error *x1)
    b_diff = sum(2*error *1)
    
    a1 = a1 - a1_diff*lr
    b = b- b_diff*lr
    print(f'{i+1}회 학습 / a1 기울기 : {a1_diff:.2f}, b기울기 : {b_diff:.2f}') # 위로 발산 하는 중 -> 러닝메이트로 훈련을 좋게
```

#### 2. sin 그래프와 가장 오차가 작은 2차 함수 그래프 그리기
- 1차 함수랑 오차가 같이나옴
- 앞에있는 계수가 거의 0 으로 만듬 1차함수보다 2차함수가 안좋아서 1차함수로 사용하는 것이 좋다

```python
import numpy as np
import matplotlib.pyplot as plt

# x 값과 sin(x) 값을 생성
x = np.arange(-3, 3, 0.01)
y = np.sin(x)

# 초기화
a = 0
b = 0
c = 0
lr = 0.000005  # 학습률

# 학습 과정
for i in range(3000):
    y_hat = a*x**2 + b*x + c  # y_hat 예측 (2차 함수)
    error = y_hat - y  # 오차 계산

    # 기울기 계산
    a_diff = sum(2 * error * x**2)
    b_diff = sum(2 * error * x)
    c_diff = sum(2 * error)

    # 파라미터 업데이트
    a = a - a_diff * lr
    b = b - b_diff * lr
    c = c - c_diff * lr

    if (i + 1) % 500 == 0:  # 500번마다 출력
        print(f'{i+1}회 학습 / a 기울기 : {a_diff:.2f}, b 기울기 : {b_diff:.2f}, c 기울기 : {c_diff:.2f}')

```

```python
plt.plot(x,y)
plt.plot(x,a*x**2 + b*x + c,'red')
plt.grid()
plt.show()
```

```python
a,b,c
# 계수를 0으로 만들어 버림
# 컴퓨터가 알아서 해준거임
```

#### 3. sin 그래프와 가장 오차가 작은 3차 함수 그래프 그리기

```python
import numpy as np
import matplotlib.pyplot as plt

# x 값과 sin(x) 값을 생성
x = np.arange(-3, 3, 0.01)
y = np.sin(x)

# 초기화
a = 0
b = 0
c = 0
d =0
lr = 0.000008  # 학습률

# 학습 과정
for i in range(3000):
    y_hat = a*x**3 + b*x**2+ c*x+d  # y_hat 예측 (2차 함수)
    error = y_hat - y  # 오차 계산

    # 기울기 계산
    a_diff = sum(2 * error * x**3)
    b_diff = sum(2 * error * x**2)
    c_diff = sum(2 * error*x)
    d_diff = sum(2 * error)

    # 파라미터 업데이트
    a = a - a_diff * lr
    b = b - b_diff * lr
    c = c - c_diff * lr
    d = d - d_diff * lr

    if (i + 1) % 500 == 0:  # 500번마다 출력
        print(f'{i+1}회 학습 / a 기울기 : {a_diff:.2f}, b 기울기 : {b_diff:.2f}, c 기울기 : {c_diff:.2f}, d 기울기 : {d_diff:.2f}')
```

```python
plt.plot(x,y)
plt.plot(x,a*x**3 + b*x**2 + c*x + d,'red')
plt.grid()
plt.show()
```

```python
a,b,c,d
```

#### 4. sin 그래프와 가장 오차가 작은 4차 함수 그래프 그리기

```python
import numpy as np
import matplotlib.pyplot as plt

# x 값과 sin(x) 값을 생성
x = np.arange(-3, 3, 0.01)
y = np.sin(x)

# 초기화
a = 0
b = 0
c = 0
d = 0
e = 0
lr = 0.0000001  # 학습률

# 학습 과정
for i in range(3000):
    y_hat = a*x**4 + b*x**3 + c*x**2 + d*x + e  # y_hat 예측 (4차 함수)
    error = y_hat - y  # 오차 계산

    # 기울기 계산
    a_diff = sum(2 * error * x**4)
    b_diff = sum(2 * error * x**3)
    c_diff = sum(2 * error * x**2)
    d_diff = sum(2 * error * x)
    e_diff = sum(2 * error)

    # 파라미터 업데이트
    a = a - a_diff * lr
    b = b - b_diff * lr
    c = c - c_diff * lr
    d = d - d_diff * lr
    e = e - e_diff * lr

    if (i + 1) % 500 == 0:  # 500번마다 출력
        print(f'{i+1}회 학습 / a 기울기 : {a_diff:.2f}, b 기울기 : {b_diff:.2f}, c 기울기 : {c_diff:.2f}, d 기울기 : {d_diff:.2f}, e 기울기 : {e_diff:.2f}')

# 결과 시각화
plt.plot(x, y, label='sin(x)')
plt.plot(x, a*x**4 + b*x**3 + c*x**2 + d*x + e, label='Fitted Quartic Function', linestyle='--')
plt.legend()
plt.show()
```

### 머신러닝

```python
import pandas as pd
import seaborn as sns
fish = pd.read_csv('fish.csv') # 도미/빙어
fish.head()
```

```python
# 길이 >25 (도미)
# 무게 <20 (빙어)

# k 최근접 이웃 분류 모델

sns.set_style('whitegrid')
sns.scatterplot(data=fish , x='length',y='weight',hue='target')
```

```python
from sklearn.neighbors import KNeighborsClassifier # KNN

X=np.array(fish[['length','weight']] )
Y=np.array(fish['target'])

model = KNeighborsClassifier() # 모델 생성
model.fit(X,Y) # 모델 학습
```

```python
a =[40,600] # a 값

model.predict([a]) # 예측 값 bream이라고 나옴

# 신뢰도 평가를 해봐야함 
```

#### 신뢰도 평가
- 데이터 비율 자동 맞춤 : train_test_split

```python
#### 신뢰도 평가 

# 훈련데이터 / 시험데이터
# train data / test data
# 최대한 훈련데이터에 쏟아야함. (100만개 -> 99만개, 1만개)
# 데이터 수가 적으면 sklearn : 75 :25
# google -> 8 : 2

# 각 세트안에 생선의 비율도 맞아야함
# 데이터를 쪼개면서 비율까지 맞춰줘야하는데,
# 비율까지 쪼개주는 라이브러리가 있다.

from sklearn.model_selection import train_test_split # 데이터 비율까지 맞춰줌

train_x,test_x, train_y,test_y = train_test_split(X,Y,test_size=0.2) # 20%를 Y 데이터에 넣어준다

```

```python
# test_x로 예측한 값과 test_y를 비교해서 정확도로 보여주는 함수
model.score(test_x,test_y)
```

#### 표준화
- 평균값이 이상치가 없어서 평균 값으로 뺀거임
- 이상치가 있으면 중앙 값으로 하는것이 좋음
- 시험데이터의 훈련데이터 평균으로 빼야함 왜냐 모르는 시험데이터의 평균으로 빼면안된다.

```python
# 표준화가 필요함
# 길이 - 길이 평균 / 길이 표준 편차 , 무게 - 무게 평균 / 무게 표준편차
```

```python
# 표준화

# 평균값이 이상치가 없어서 평균 값으로 뺀거임
# 이상치가 있으면 중앙 값으로 하는것이 좋음

# 중앙값을 빼는게 로버스트?스케일링이라고함

mean=train_x.mean(axis=0) # 열방향으로 평균이 구해짐
std = train_x.std(axis=0) # 

train_scaled = (train_x - mean)/std
test_scaled = (test_x - mean)/std # 시험데이터의 훈련데이터 평균으로 빼야함 왜냐 모르는 시험데이터의 평균으로 빼면안된다.
```

```python
plt.scatter(train_scaled[:,0],train_scaled[:,1]) #훈련 데이터
plt.show()
```

```python
model=KNeighborsClassifier()
model.fit(train_scaled,train_y) # 표준화된 훈련데이터로 학습했고
model.score(test_scaled,test_y) # 표준화된 test로 예측함
```

#### 모델링 , 신뢰도 평가, 표준화 , 이웃 수 score 측정 후 결정

```python
from sklearn.model_selection import train_test_split # 데이터 비율까지 맞춰줌
X=fish2.iloc[:,1:].to_numpy()
Y=fish2['Species'].to_numpy()

train_x,test_x, train_y,test_y = train_test_split(X,Y) # 이렇게 하면 기본 값으로 됨 75 : 25

mean=train_x.mean(axis=0) # 열방향으로 평균이 구해짐
std = train_x.std(axis=0) # 

train_scaled = (train_x - mean)/std
test_scaled = (test_x - mean)/std

model=KNeighborsClassifier() # 이웃수 : 5
model.fit(train_scaled,train_y) # 표준화된 훈련데이터로 학습했고

box = []
for n in range(1,31):
    model.n_neighbors = n # 이웃 숫자를 바꿈
    score = model.score(test_scaled,test_y)
    box.append(score) # 이웃숫자를 30개를 넣어서 확인 해봄
```

```python
plt.plot(range(1,31),box) # 이웃숫자의 신뢰도 평가 
plt.show() # 1개로 했을 때 젤 좋게 나옴

model.n_neighbors =1
```

- k 최근접 이웃 회귀모델
- 공공자전거 수요

```python
import pandas as pd
from sklearn.neighbors import KNeighborsRegressor
bike = pd.read_csv('bike data.csv')
bike.head()
del bike['datetime']

# sns.scatterplot(data=bike, x='windspeed',y='count')

# # 0인 값이 너무 많다 풍속을 측정하는 기기가 고장난다고 생각하고 제거하고 하는 것이 성능이 더 좋았다.

X = bike.iloc[:,:-1].to_numpy()
Y = bike['count'].to_numpy()


train_x,test_x, train_y,test_y = train_test_split(X,Y) # 이렇게 하면 기본 값으로 됨 75 : 25

mean=train_x.mean(axis=0) # 열방향으로 평균이 구해짐
std = train_x.std(axis=0) # 

train_scaled = (train_x - mean)/std
test_scaled = (test_x - mean)/std

model=KNeighborsRegressor() 
model.fit(train_scaled, train_y)
model.score(test_scaled, test_y) # 회귀모델에서는 : 결정계수이다. 예측값의 분산 / 실제값의 분산


# 예측값이 실제값을 28% 만큼 설명한다
```

```python
# r^2 , MAE , RMSE

pred = model.predict(test_scaled)

abs(pred - test_y).mean() # 평균적으로 108대 정도는 틀린다
```

```python
(((pred - test_y)**2).mean())**(1/2) # RMSE
```

```python
iris = sns.load_dataset('iris') # 세토시,버지컬러, 버지니카 -> 꽃 데이터
iris.head()
```

```python
iris['species'].value_counts()
```

```python
X = iris.iloc[:,:-1].to_numpy()
Y = iris['species'].to_numpy()

train_x, test_x , train_y , test_y = train_test_split(X,Y)
```

```python
from sklearn.tree import DecisionTreeClassifier,plot_tree # 분류트리 모델
# plot_tree 결정 트리모델 시각화 해줌

model = DecisionTreeClassifier(max_depth=2)
# max_depth=2 깊이 조절을 해줌으로써 과적합을 방지할 수 있다.
model.fit(train_x,train_y)
model.score(test_x,test_y)

print(model.score(train_x,train_y)) # 훈련데이터
print(model.score(test_x,test_y)) # 시험 데이터

# 훈련데이터와 시험데이터가 차이가 잘 안나야 과적합이 안난 것이다.

# 훈련 >>>>> 시험 : 과대적합
# 훈련 > 시험 : 정상
# 훈련 < 시험 : 과소적합 의심을 해봐야함
# 둘다 작게 나오는 경우 : 과소적합

# 결정트리는 : score를 훈련,시험 둘다 평가하는 것이 좋다.
```

```python
plt.figure(figsize = (10, 3))
plot_tree(model, filled = True, feature_names=['sl', 'sw', 'pl', 'pw'])
plt.show()
```
